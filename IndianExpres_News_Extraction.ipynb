```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

# importing modules
import urllib.request

## To extract Business Standard News
# url='https://indianexpress.com/section/political-pulse/'
# url = 'https://indianexpress.com/page/'
url = 'https://community.kobotoolbox.org/'
## agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
agent = {'User-Agent': 'Mozilla/5.0'}
page=requests.get(url,headers=agent)
soup=BeautifulSoup(page.text,'html.parser')
```

```python
## Extracting page links
page_links=[]

for page_num in range(2, 5):
  # page_url = url + '/page/' + str(page_num) + "/"
  page_url = url + str(page_num) + '/?s=population+control'
  page_links.append(page_url)

page_links[0:2]
```

```python
import urllib.request

str_for_search = []

opener = urllib.request.build_opener()
opener.addheaders = [('User-Agent', 'MyApp/1.0')]
urllib.request.install_opener(opener)

for url in page_links:
  html = urllib.request.urlopen(url)
  htmlParse = BeautifulSoup(html, 'html.parser')

  # para = htmlParse.find_all('div', class_='list-news')
  # # para_ = para.find('a')
  # print(para)
  # # str_for_search.append(para_)

  for para in htmlParse.find_all('div', class_='details'):
      para_ = [a['href'] for a in para.find_all('a', href=True) if a.text]
      print(para_[0])
      str_for_search.append(para_[0])
```
